```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('data/sells.xlsx')

df.head()
```

```python
categories = pd.read_excel('data/product_categories.xlsx')

df = pd.merge(df, categories, how='inner', on='Number_Category')

df.head()
```

```python
df['Product_Price'] = df['Sale_Amount'] / df['Quantity']

df.head()
```

```python
print(f"Number of unique regions: {len(df['Region'].unique())}\n"
      f"Number of unique markets: {len(df['Market'].unique())}\n"
      f"Number of unique states: {len(df['State'].unique())}\n")
```

```python
df['Quantity'].groupby(df['Region']).sum()
```

```python
df['Sale_Amount'].groupby(df['Region']).sum()
```

```python
df['Quantity'].groupby(df['State']).sum()
```

```python
df['Sale_Amount'].groupby(df['State']).sum()
```

```python
df.groupby('Region')['Sale_Amount'].agg(
    mean='mean',
    max='max',
    min='min',
    sum='sum',
    mode=lambda x: x.mode().iloc[0],
    median='median',
    var='var')
```

```python
plt.xlabel('Quantity', fontsize=20)
plt.ylabel('Count', fontsize=20)
df['Quantity'].hist(bins=5,
                    grid=False,
                    range=(0, 70),
                    rwidth=0.8,
                    ylabelsize=20,
                    xlabelsize=20,
                    figsize=(12, 8),
                    color='orange',
                    edgecolor='black')
```

```python
plt.figure(figsize=(20, 8))
plt.title('Quantity mean by Product Category', fontsize=30)
plt.xlabel('Product_Category', fontsize=20)
plt.ylabel('Quantity mean', fontsize=20)
sns.barplot(y='Quantity',
            hue='Product_Category',
            legend=True,
            palette='Set1',
            gap=0.5,
            data=df)
```
